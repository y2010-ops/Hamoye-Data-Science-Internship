```python
# Q.8,Q.4
import pandas as pd
url= 'https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')



```

```python
# Q.5
print(fuel_data['fuel_qty_burned'].skew())
print(fuel_data['fuel_qty_burned'].kurt())
```

```python
print(fuel_data.dropna())
print(fuel_data.dropna().count())
print(fuel_data.dropna().sum())
print(fuel_data.fillna("NA"))
```

```python
# Q.10
df = pd.DataFrame(fuel_data)
print(df['fuel_cost_per_unit_delivered'].idxmax())
print(df.iloc[3564])
```

```python
# Q.6
y=df.dropna().count()
x= df.count()
print(y)
print(x)
Rows_missing= x-y
print(Rows_missing)
percentage= (y/x)*100
print(percentage)
print(df.isnull().sum())
print(df.isnull().sum().sum())
```
